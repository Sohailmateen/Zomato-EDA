* The Country-Code(df_country) dataset and zomato.csv(df) dataset both have the feature named as Country Code
* So We can merge them (left join)

```python
final_df = pd.merge(df, df_country, on='Country Code',how='left')
```

### Observation: Based on the number of restaurant records in the dataset, India has the maximum number of transactions/records ($8,652$) (94.39%), followed by the United States ($434$)(4.7%) and the United Kingdom ($80$)(0.887%).

```python
final_df.columns
```

```python
ratings = final_df.groupby(['Aggregate rating','Rating color', 'Rating text']).size().reset_index().rename(columns={0:'Rating Count'})
```

```python
ratings
```

* The most striking observation is the high volume of unrated restaurants (2,148 records with a 0.0 rating).

* Among the restaurants that are rated, the distribution is:

* Most Common: The majority of rated restaurants cluster in the Orange (Average) range (2.5 to 3.4).

* Highest Count Point: 3.2 (Average) has the highest count of any single-rated point (522 records).

* Least Common (Rated): Ratings classified as Dark Green (Excellent) (4.5 to 4.9) and Red (Poor) (1.8 to 2.4) have the lowest counts.

| Rating Text | Rating Range | Total Count (Approximate Sum) | Interpretation |
|--------------|--------------|-------------------------------|----------------|
| Not rated | 0.0 | 2,148 | The largest single group; lacks data. |
| Poor | 1.8 - 2.4 | 186 | Very few restaurants fall into the lowest tier. |
| Average | 2.5 - 3.4 | 3,674 | The largest group of rated restaurants. |
| Good | 3.5 - 3.9 | 2,100 | High volume of moderately high-quality establishments. |
| Very Good | 4.0 - 4.4 | 1,079 | Substantial number of highly-rated restaurants. |
| Excellent | 4.5 - 4.9 | 246 | The smallest group of rated restaurants. |


```python
ratings.head()
```

```python
plt.rcParams['figure.figsize'] = (12, 6)
sns.barplot(
    data=ratings,
    x='Aggregate rating',
    y='Rating Count',
    hue='Rating color',
    palette=['wheat','red','orange','yellow','green','green']
    # palette='viridis'   # Try 'coolwarm', 'magma', 'Set2', 'Spectral','viridis', etc.
)
plt.title('Aggregate Rating vs Rating Count', fontsize=14)
plt.xlabel('Aggregate Rating')
plt.ylabel('Rating Count')
plt.show()
```

#### Obsersvation
    1. Not Rated count is very high
    2. Maximum number of ratings are in between 2.5 to 4.1

```python
### Count plot
sns.countplot(
    data=ratings,
    x='Rating color',
    hue='Rating color',
    palette=['wheat','red','orange','yellow','green','green']
)
```
